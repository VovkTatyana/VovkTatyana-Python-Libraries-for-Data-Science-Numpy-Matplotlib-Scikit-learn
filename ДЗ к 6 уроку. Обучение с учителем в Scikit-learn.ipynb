Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки

Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.
Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
```

```python
boston_house = load_boston()
boston_house.keys()
```

```python
data = boston_house["data"]
data.shape
```

```python
feature_names = boston_house["feature_names"]
feature_names
```

```python
print(boston_house["DESCR"])
```

```python
target = boston_house["target"]
target[:10]
```

```python
X = pd.DataFrame(data, columns=feature_names)

X.head(10)
```

```python
X.info()
```

```python
y = pd.DataFrame(target, columns=["price"])
y.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, shuffle=True, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred.shape
```

```python
check_test = pd.DataFrame({
 "y_test": y_test["price"],
 "y_pred": y_pred.flatten(),
})
check_test.head()
```

```python
from sklearn.metrics import r2_score
```

```python
check_test["error"] = check_test["y_pred"] - check_test["y_test"]
check_test.head()
```

```python
R2_1 = r2_score(check_test["y_pred"], check_test["y_test"])
R2_1
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
```

```python
df = pd.read_csv('C://Users//User//Desktop//creditcard.csv')
df.head()
```

```python
df["Class"].value_counts(normalize=True)
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
```

```python
df.head(10)
```

```python
X = df.drop("Class", axis=1)
X.head()
```

```python
y = df["Class"]
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
print('X_train ', X_train.shape)
print('X_test ', X_test.shape)
print('y_train ', y_train.shape)
print('y_test ', y_test.shape)
```

```python
parameters = [{'n_estimators': [10, 15], 'max_features': np.arange(3, 5), 'max_depth': np.arange(4, 7)}]
```

```python
model_1 = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc', cv=3)
```

```python
model_1.fit(X_train, y_train)
```

```python
model_1.best_params_
```

```python
model_1 = RandomForestClassifier(max_depth=6, max_features=3, n_estimators=15)

model_1.fit(X_train, y_train)
```

```python
y_pred = model_1.predict_proba(X_test)
```

```python
y_pred_proba = y_pred[:, 1]
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred_proba)
```
